```python
csv_1.set_index('human_time')
```

```python
csv_1.plot(x='human_time', y='web_cpu', kind='scatter')
```

```python
csv_1.plot(x="epoch_time", y="nginx_cpu", kind='scatter')
```

```python
x = csv_1["epoch_time"]
y1 = csv_1["web_cpu"]
y2 = csv_1["mongo-db_cpu"]
y3 = csv_1["nginx_cpu"]
y4 = csv_1['mosquitto_cpu']
y5 = csv_1['host_cpu']
fig, ax = plt.subplots()
ax.scatter(x, y1, label="web_cpu")
ax.scatter(x, y2, label='mongo_cpu')
ax.scatter(x, y3, label='nginx_cpu')
ax.scatter(x, y4, label='mqtt_cpu')
ax.scatter(x, y5, label='host_cpu')
ax.legend()
fig.savefig(f'{files[0].split(".")[0]}.png', dpi=300)
```
